```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
phi = lambda t : 1 if (t<=1 and t>=0) else 0
phi22 = lambda t : 2*phi(4*t-2)
phi42 = lambda t : 4*phi(16*t-2)
f = lambda t : np.sin(2*np.pi*t)
```

```python
def P(f,t,j):
    n=2**j-1
    p=0
    for k in range(0,n):
        Cjk=2**(j/2)*integrate.quad(f,k*2**(-j),(k+1)*2**(-j))[0]
        phijk= lambda x : 2**(j/2)*phi(2**j*x-k)
        p+=Cjk*phijk(t)
    return p    
        
```

```python
X=np.arange(-0.5,1.5,0.01)
X1=np.arange(0.1,1,0.01)
App3=[P(f,x,3) for x in X1]
App5=[P(f,x,5) for x in X1]
Ex=[f(y) for y in X1]
Y=[phi(x) for x in X]
Y1=[phi22(x) for x in X]
Y2=[phi42(x) for x in X]
plt.subplot(131)
plt.plot(X,Y)
plt.plot(X,Y1)
plt.plot(X,Y2)
plt.subplot(132)
plt.plot(X1,App3)
plt.plot(X1,Ex)
plt.subplot(133)
plt.plot(X1,App5)
plt.plot(X1,Ex)
plt.show()
```
