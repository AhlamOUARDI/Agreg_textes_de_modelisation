# Propagation d'un virus 

```python
import numpy as np
import matplotlib.pyplot as plt
```

  **Modèle proportionnel**

```python
def trajectoire(X_0,m,alpha,p=1/5):
    P=[1-p,p]
    U=np.random.choice([0,1],m,p=P)
    X=[X_0]
    for i in range (0,m-1):
        U_=U[i]
        X.append((1+alpha)**U_*(1-alpha)**(i-U_)*X_0)
    s=0
    for i in X:
        if i <X_0:
            s+=1
        s=s/len(X)
    print('P( X < X0)=', s)
    return X
```

```python
alpha = 0.5
n=20
p=0.5
X=trajectoire(30,n,alpha,p)
X1=trajectoire(50,n,alpha,p)
X2=trajectoire(80,n,alpha,p)
plt.xlabel('L\'instant t')
plt.ylabel('Nmbre de personnes malades à t')
plt.plot(np.arange(n),X,label='X_0=30')
plt.plot(np.arange(n),X1,label='X_0=50')
plt.plot(np.arange(n),X2,label='X_0=80')
plt.legend()
```
