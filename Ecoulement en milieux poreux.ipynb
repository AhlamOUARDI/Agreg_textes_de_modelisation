```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g   = lambda t : 10 if t<=0.8 and 0.5<= t else 0
phi = lambda t : abs(t)*t
phii= lambda t : np.sqrt(abs(t)) if t>=0 else -np.sqrt(abs(t))
N=100
h=1/(N+1)
def dich(alpha,x):
    p=[0]
    s=[0]
    for i in range (N):
        s.append (s[-1]+h*g(x[i+1]))
        p.append (p[-1]+h*phii(phi(alpha-s[i])))
    p.append (p[-1]+h*phii(phi(alpha-s[i])))
    return p
        
```

```python
X=np.linspace(0,1,102)
plt.plot(X,dich(0,X))
plt.plot(X,dich(0.4,X))
plt.plot(X,dich(0.8,X))
plt.plot(X,dich(1.2,X))
plt.plot(X,dich(1.6,X))
plt.plot(X,dich(2,X))
```
