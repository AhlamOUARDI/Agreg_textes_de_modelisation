# Interpolation et krigeage

**Objectif** : prévoir la valeur prise par un phénomène naturel en un site à partir des observations ponctuelles en des sites voisins

Cherchons $\hat u$ de la forme $$\hat u (x) = d(x)+\omega(x)$$
avec $$d(x)= \Sigma_{i=0}^{k-1} a_i p_i(x) \text{   , et    }\omega(x)= \Sigma_{j=1}^{n} \alpha_j \phi(|x-x_j|).$$

Alors les coefficients sont solutions du système linéaire suivant : 
$$\begin{pmatrix}
C & P\\
^t P & 0
\end{pmatrix} \begin{pmatrix}
\alpha\\
a
\end{pmatrix} = \begin{pmatrix}
u\\
0
\end{pmatrix}$$

**Spline cubique** :
Montrons que la fonction obtenue par krigeage en choisissant une dérive affine et une fonction $\phi(h) = h^3$ est une spline cubique :

Soit $$\hat u (x) = a_0+a_1 x+\Sigma_{j=1}^{n} \alpha_j |x-x_j|^3$$
$u$ est de Classe $C^2$

$$
\phi(|x-x_j|) = \left\{
    \begin{array}{ll}
        (x-x_j)^3 & \mbox{si } x_j \leq x \\
        -(x-x_j)^3 & \mbox{sinon.}
    \end{array}
\right.
$$
donc $$
\phi ' (|x-x_j|) = \left\{
    \begin{array}{ll}
        3(x-x_j)^2 & \mbox{si } x_j \leq x \\
        -3(x-x_j)^2 & \mbox{sinon.}
    \end{array}
\right.
$$
et la limite à gauche et à droite en $x_j$ égale à 0, donc $\phi$ est dérivable.
$$
\phi '' (|x-x_j|) = \left\{
    \begin{array}{ll}
        6(x-x_j) & \mbox{si } x_j \leq x \\
        -6(x-x_j) & \mbox{sinon.}
    \end{array}
\right.
$$ de même la limite à gauche et à droite en $x_j$ égale à 0, donc $\phi$ est deux fois dérivable et de degré 3.

Alors $u$ est une spine cubique.

Montrons que c'est naturelle, i.e. $u''(x_1)=u''(x_n)=0$

$$u''(x_1)=-6\Sigma_{j=1}^{n} \alpha_j (x_1-x_j)=-6x_1\Sigma_{j=1}^{n} \alpha_j  +6\Sigma_{j=1}^{n} \alpha_j x_j=0$$

$$u''(x_n)=-6\Sigma_{j=1}^{n} \alpha_j (x_n-x_j)=-6x_n\Sigma_{j=1}^{n} \alpha_j  +6\Sigma_{j=1}^{n} \alpha_j x_j=0$$

car $\forall i \in [|0,k-1|]$, $\Sigma_{j=1}^{n} \alpha_j p_i(x_j)=0$ avec $p_j(x)=x^j$

Alors $u$ est une spine cubique naturelle.



```python
import numpy as np
import matplotlib.pyplot as plt
phi1 = lambda h: h
phi2 = lambda h: h**3
phi3 = lambda h: np.exp(-h**2)
```

```python
#C
def C(x, phi):
    n = len(x)
    C = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            C[i][j] = phi(abs(x[i]-x[j]))
    return C

#P
def P(x, k): 
    n = len(x)
    P = np.zeros((n, k))
    for i in range(n):
        for j in range(k):
            P[i,j] = x[i]**j
    return P

#M
def M(x, phi,C, k):
    n = len(x)
    M = np.zeros((n+k, n+k))
    M[0 : n , 0 : n]  = C(x,phi)
    M[0 : n , n : n+k]= P(x,k)
    M[n:n+k , 0 : n]  = P(x,k).T
    return M

#alpha & a
def coeff(M,U):
    u = np.zeros(len(M))
    u[:len(U)] = U
    X = np.linalg.solve(M, u)
    alpha = X[:len(U)]
    a = X[len(U):]
    return alpha, a

# u
def u(X, x, U, phi,k):
    def d(x,a):
        s=0.
        for i in range (0,len(a)):
            s+= a[i]*x**i
        return s
    def w(X,x,alpha, phi):
        s=0.
        for i in range (len(alpha)):
            s+= alpha[i]*phi(abs(x-X[i]))
        return s
    Mat = M(x, phi, C,k)
    alpha, a = coeff(Mat, U)
    return d(X, a) + w(x, X, alpha, phi)
```

**Effet pépite :**

L'idée c'est de diminuer l'influence d'une donnée marginale ou aberrante en ajoutant à la matrice $C$ une matrice diagonale $D$ avec $d_{i,i}$ d’autant plus grand qu'on veut moins tenir compte du point $(x_i,u_i)$ dans l'interpolation. Lorsque l'effet pépite devient important, on obtient un problème de moindres carrés pondérés. Plus précisément, on ajoute à $C$ une matrice diagonale $\beta D$ et on fait tendre $\beta$ vers $\infty$ . Le système s’écrit alors
$$
 \left\{
    \begin{array}{ll}
        (C+\beta D) \alpha +P a & = U \\
        ^tP \alpha &  = 0
    \end{array}
\right.
$$
Pour $\beta$ assez grande, $C+\beta D$ est inversible, car à diag strictement dominante. Ce qui implique $$^t P(C+\beta D )^{-1} P a = ^t P(C+\beta D )^{-1} U$$

**Proof Thm1** :
lorsque $\beta \rightarrow \infty$, la solution tend vers celle de $$^tP D^{-1} P a = ^tP D^{-1} U$$
On a
$$^t P(\beta D(\frac{1}{\beta}D^{-1}C+I) )^{-1} P a = ^t P(\beta D(\frac{1}{\beta}D^{-1}C+I) )^{-1} U$$
On a 
$\rho (\frac{1}{\beta}D^{-1}C ) < 1$ car $\beta \rightarrow \infty$, donc $\frac{1}{\beta}D^{-1}C+I$ est inversible, 

Avec $$(I+\frac{1}{\beta}D^{-1}C)^{-1}=\Sigma_{k=0}^\infty(\frac{-1}{\beta}D^{-1}C)^k= I + \Sigma_{k=1}^\infty(\frac{-1}{\beta}D^{-1}C)^k \rightarrow I$$

En remplaçant $(I+\frac{1}{\beta}D^{-1}C)^{-1}$ par $I$, on obtient : $$^t PD^{-1}Pa=^tPD^{-1}U$$
