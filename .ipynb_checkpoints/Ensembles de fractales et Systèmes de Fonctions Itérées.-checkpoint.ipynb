```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1= lambda x,y : np.array([0.5*x,0.5*y])+np.array([0,0.5])
w2= lambda x,y : np.array([0.5*x,0.5*y])+np.array([-0.5,-0.5])
w3= lambda x,y : np.array([0.5*x,0.5*y])+np.array([0.5,-0.5])
```

```python
def frac(a):
    x=np.linspace(-1,1,5)
    y=np.linspace(-1,1,5)
    L=[[i,j] for i in x for j in y]
    k=1
    while k<=a:
        L_=L
        L=[]
        x=[L_[i][0] for i in range (len(L_))]
        for i in range (len(x)) :
            L.append(w1(L_[i][0],L_[i][1]))
            L.append(w2(L_[i][0],L_[i][1]))
            L.append(w3(L_[i][0],L_[i][1]))
        L_=[]
        k+=1
    return L
```

```python
L=frac(3)
X=[L[i][0] for i in range (len(L))]
Y=[L[i][1] for i in range (len(L))]
plt.scatter(X,Y)
```

```python
w1= lambda x,y : np.array([0.5*x,y])
w2= lambda x,y : np.array([0.5*x,y])+np.array([0.5,0])
def tric(a):
    x=np.linspace(0,0.5,20)
    y=np.linspace(0.5,1,20)
    L=[[i,2*i] for i in x]+[[i,-2*i+2] for i in y]
    k=1
    while k<=a:
        L_=L
        L=[]
        x=[L_[i][0] for i in range (len(L_))]
        for i in range (len(x)) :
            L.append(w1(L_[i][0],L_[i][1]))
            L.append(w2(L_[i][0],L_[i][1]))
        L_=[]
        k+=1
    return L
```

```python
L=tric(5)
X=[L[i][0] for i in range (len(L))]
Y=[L[i][1] for i in range (len(L))]
plt.scatter(X,Y)
```

```python
d=0.5
def w(n,x,y,a,c,d,e,f):
    return np.dot(np.array([[a,0],[c,d]]),np.array([x,y]))+np.array([e,f])
N=3
x=[0,1/2,1]
y=[1/2,2,1/4]
A=[(x[i-1]-x[i])/(x[0]-x[-1]) for i in range(1,N)]
E=[(x[i]*x[0]-x[i-1]*x[-1])/(x[0]-x[-1]) for i in range(1,N)]
C=[(y[i-1]-y[i]-d*(y[0]-y[-1]))/(x[0]-x[-1]) for i in range(1,N)]
F=[(y[i]*x[0]-y[i-1]*x[-1]+d*(y[0]*x[-1]-x[0]*y[-1]))/(x[0]-x[-1]) for i in range(1,N)]

W=[w(i,x[k],y[k],A[i],C[i],d,E[i],F[i]) for i in range(N-1) for k in range(N)]
n=len(W)
X=[W[i][0] for i in range(n)]
Y=[W[i][1] for i in range(n)]
plt.scatter(X,Y)
plt.plot(x,y)
plt.show()
```

```python
d=0.02
w1= lambda x,y : np.array([0.5*x,(1.625+d)*y])+np.array([0,0.25])
w2= lambda x,y : np.array([0.5*x,(-1.625+d)*y])+np.array([0.5,1.75])
def Int(a):
    x=[0,1/2,1]
    y=[1/2,2,1/4]
    L=[[x[i],y[i]] for i in range (3)]
    k=1
    while k<=a:
        L_=L
        L=[]
        x=[L_[i][0] for i in range (len(L_))]
        for i in range (len(x)) :
            L.append(w1(L_[i][0],L_[i][1]))
            L.append(w2(L_[i][0],L_[i][1]))
        L_=[]
        k+=1
    return L
```

```python
L=Int(0)
X=[L[i][0] for i in range (len(L))]
Y=[L[i][1] for i in range (len(L))]
plt.plot(X,Y)
```
