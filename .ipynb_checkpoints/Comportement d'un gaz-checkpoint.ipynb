**I. Introduction**

On répartit $2n$ boules, n noires et n blanches, dans 2 urnes $U_1$ et $U_2$ à raison de n boules par urne. Puis à chaque instant on choisit une boule au hazard dans chacune des urnes et on les échange. On désigne par $X_n$ le nombre de boules noires dans l'urne $U_1$ après n tirages. \\
La suite $X = {X_n; n \in \mathbb N}$ est une chaîne de Markov à valeurs dans l'espace d'état $E = {0, 1, . . . , n}$

**II. Aspect mathématique**

**Prop 1.** Soit une chaîne de Markov homogène $X \sim(E, \mu, K)$.\\
Alors $\forall n \geq 0, \forall i_0, i_1, \ldots, i_n \in E$
$$
\mathbb{P}\left(X_0=i_0, X_1=i_1, \ldots, X_n=i_n\right)=\mu\left(i_0\right) K\left(i_0, i_1\right) K\left(i_1, i_2\right) \cdots K\left(i_{n-1}, i_n\right)$$


**Pve :** Soit $n \in \mathbb{N}^*, i_0, \ldots, i_n \in E$. Si $\mathbb{P}\left(X_0=i_0, \ldots, X_{n-1}=i_{n-1}\right)>0$, une utilisation successive de la formule des probabilités conditionnelles donne
$$
\begin{aligned}
\mathbb{P}\left(X_0=i_0, \ldots, X_n\right.  \left.=i_n\right) 
& =\mathbb{P}\left(X_0=i_0\right) \mathbb{P}\left(X_1=i_1 \mid X_0=i_0\right) \cdots \mathbb{P}\left(X_n=i_n \mid X_0=i_0, \ldots, X_{n-1}=i_{n-1}\right) \\
& =\mu_0\left(i_0\right) \prod_{k=1}^n K\left(i_{k-1}, i_k\right) .
\end{aligned}
$$
Si $\mathbb{P}\left(X_0=i_0, \ldots, X_{n-1}=i_{n-1}\right)=0$, soit $\mathbb{P}\left(X_0=i_0\right)=0$ i.e. $\mu_0\left(i_0\right)=0$ et donc la prop reste vrai; soit il existe $m \in\{1, \ldots, n-1\}$ tel que $\mathbb{P}\left(X_0=i_0, \ldots, X_{m-1}=x_{m-1}\right)>0$ et $\mathbb{P}\left(X_0=i_0, \ldots, X_m=\right.$ $\left.i_m\right)=0$. Dans ce dernier cas, si $n=m$, on obtient que
$$
0=\mathbb{P}\left(X_0=i_0, \ldots, X_m=i_m\right)=\mu_0\left(i_0\right) \prod_{k=1}^m K\left(i_{k-1}, i_k\right)
$$

**Prop 3.** Soit une chaîne de Markov homogène $X \sim(E, \mu, K)$. Alors
$$
\forall n \geq 0, \forall m>0, \forall i, j \in E, \quad \mathbb{P}\left(X_{n+m}=j / X_n=i\right)=\mathbb{P}\left(X_m=j / X_0=i\right)=K^{(m)}(i, j)
$$

**Pve :**
Calculons $\mathbf{P}\left(\mathrm{X}_{n+m}=j \mid \mathrm{X}_0=i\right)$ en décomposant l'événement $\left\{\mathrm{X}_{n+m}=j\right\}$ sur les événements $\left(\left\{\mathrm{X}_n=k\right\}\right)_{k \in \mathrm{E}}$ qui forment une partition de $\Omega$ :
$$
\begin{aligned}
\mathbf{P}\left(\mathrm{X}_{n+m}=j \mid \mathrm{X}_0=i\right) & =\mathbf{P}\left(\mathrm{X}_{n+m}=j \cap\left(\cup \mathrm{X}_n=k\right) \mid \mathrm{X}_0=i\right) \\
& =\sum_{k \in \mathrm{E}} \mathbf{P}\left(\mathrm{X}_{n+m}=j \cap \mathrm{X}_n=k \mid \mathrm{X}_0=i\right) \\
& =\sum_{k \in \mathrm{E}} \frac{\mathbf{P}\left(\mathrm{X}_{n+m}=j, \mathrm{X}_n=k, \mathrm{X}_0=i\right)}{\mathbf{P}\left(\mathrm{X}_0=i\right)} \\
& =\sum_{k \in \mathrm{E}} \frac{\mathbf{P}\left(\mathrm{X}_{n+m}=j \mid \mathrm{X}_n=k, \mathrm{X}_0=i\right) \mathbf{P}\left(\mathrm{X}_n=k, \mathrm{X}_0=i\right)}{\mathbf{P}\left(\mathrm{X}_0=i\right)} \\
& =\sum_{k \in \mathrm{E}} \mathbf{P}\left(\mathrm{X}_m=j \mid \mathrm{X}_0=k\right) \mathbf{P}\left(\mathrm{X}_n=k \mid \mathrm{X}_0=i\right)
\end{aligned}
$$
car $\mathrm{X}_n$ est une chaîne de Markov homogène.
En particulier, pour $n=m=1$ et pour tous $(i, j) \in \mathrm{E} \times \mathrm{E}$,
$$
\mathbf{P}\left(\mathrm{X}_2=j \mid \mathrm{X}_0=i\right)=\sum_{k \in \mathrm{E}} \mathbf{P}\left(\mathrm{X}_1=j \mid \mathrm{X}_0=k\right) \mathbf{P}\left(\mathrm{X}_1=k \mid \mathrm{X}_0=i\right)=\sum_{k \in \mathrm{E}} \mathrm{P}(i, k) \mathrm{K}(k, j)
$$
ce qui établit la relation dans le cas $n=2$. Pour $m=1$ et $n>1$, On a $$
\begin{aligned}
\forall(i, j) \in \mathrm{E} \times \mathrm{E} \quad \mathbf{P}\left(\mathrm{X}_{n+1}=j \mid \mathrm{X}_0=i\right) & =\sum_{k \in \mathrm{E}} \mathbf{P}\left(\mathrm{X}_n=k \mid \mathrm{X}_0=i\right) \mathbf{P}\left(\mathrm{X}_1=j \mid \mathrm{X}_0=k\right) \\
& =\sum_{k \in \mathrm{E}} \mathbf{P}\left(\mathrm{X}_n=k \mid \mathrm{X}_0=i\right) \mathrm{K}(k, j)
\end{aligned}
$$
donc si $\mathbf{P}\left(\mathrm{X}_n=k \mid \mathrm{X}_0=i\right)=\mathrm{K}^n(i, k)$ alors $\mathbf{P}\left(\mathrm{X}_{n+1}=j \mid \mathrm{X}_0=i\right)=\mathrm{K}^{n+1}(i, j)$ et la récurrence est établie.


**Prop 4.** Soit une chaîne de Markov homogène $X \sim(E, \mu, K)$. Nous avons alors les équations de Chapman-Kolmogorov suivantes :
$$
\forall m \geq 2, \forall 0<r<m, \forall i, j \in E, \quad K^{(m)}(i, j)=\sum_{k \in E} K^{(m-r)}(i, k) K^{(r)}(k, j)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean
```

```python
n=5 #nbre de boules noires
#la matrice de transition K
K=np.diag([(i/n)**2 for i in range(1,n+1)],-1)+np.diag([2*i*(n-i)/n**2 for i in range(n+1)])+np.diag([((n-i)/n)**2 for i in range(0,n)],1)
[sum(P[k,:]) for k in range (n)]
```

```python
K
```

```python
#Trajectoire de longueur m
def traj(m,M,X0):
    X=[X0]
    for k in range (0,m):
        X.append(np.random.choice(range(n+1),p=M[X[k],:]))
    return X
```

```python
X0=5
m=20
plt.plot(np.arange(m+1),traj(m,P,X0))
plt.ylabel('Nbre de boules noires')
plt.xlabel('Après i tirage')
```

```python
X0=5
m=100000
T=traj(m,P,X0)
plt.hist(T)
```

```python
plt.plot(np.arange(m+1),T)
```

```python
#Une approximation de π_k pour les différentes valeurs de k
def Pi(k):
    pi=0
    for i in range (1,m+1):
        if T[i]==k:
            pi+=1
    pi=pi/m        
    return pi
```

```python
print ( 'Pour k=0:', Pi(0))
print ( 'Pour k=1:', Pi(1))
print ( 'Pour k=2:', Pi(2))
print ( 'Pour k=3:', Pi(3))
print ( 'Pour k=4:', Pi(4))
print ( 'Pour k=5:', Pi(5))
```

```python
sum([Pi(i) for i in range (6)])
```

```python
#l'évolution du nombre moyen de boules noires dans l'urne U_2
nbreMoy=[mean(traj(m,P,X0)) for m in [10,20,30,50,100,200]]
plt.plot([10,20,30,50,100,200],nbreMoy,'o')
plt.xlabel('m')
plt.ylabel('Nbre moyen de bouls noires dans $U_2$')
```
