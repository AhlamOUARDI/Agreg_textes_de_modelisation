```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g1 = lambda u : u
g2 = lambda u : u*(1-u)
g3 = lambda u : u*(1-u**2)
def Euler(u0,L,N,g):
    h=L/N
    U=[u0]
    for i in range(N) :
        U.append(U[-1]+h*g(U[-1]))
    return U
```

```python
U0=0.1
b=10
N=100
Y1=Euler(U0,2.5,N,g1)
Y2=Euler(U0,b,N,g2)
Y3=Euler(U0,b,N,g3)
X1 = np.linspace(0,2.5,N+1)
X = np.linspace(0,10,N+1)
plt.plot(X1,Y1,"-.", label='u')
plt.plot(X,Y2,"--", label='u(1-u)')
plt.plot(X,Y3, label='$u(1-u^2$)')
plt.legend()
```

```python
L=40
J=50
tau=0.01
h=2*L/(J-1)
a = tau/h**2
M=(1+2*a)*np.eye(J) -a*np.eye(J,k=-1)-a*np.eye(J,k=+1)
M[0,0]=1+a
M[J-1,J-1]=1+a
X=np.linspace(-L,L,J)
print(M)
print(a)
```

```python
T=5
N=int(T//tau)
Ci= lambda t : 1/2+1/2*np.tanh(t/(2*sqrt(2)))
Op= lambda t,x : 1/2+1/2*np.tanh((1/(2*sqrt(2)))*(x+3*t/sqrt(2)))
U = np.array([Ci(x) for x in X ])
for i in range(N):
    U = np.linalg.solve(M,U+tau*g3(U))


Y=Op(T,X)
plt.plot(X,Y, label='Onde progressive')    
plt.plot(X,U,"b--+", label='Schma numrique')
plt.xlabel('x')
plt.title('T=5')
plt.ylabel('U')
plt.legend()
```

```python
U0 = lambda x : 0.8*np.cos(np.pi*x/20) if abs(x)<=10 else 0
```

```python
T=1
N=int(T//tau)
U = np.array([U0(x) for x in X ])
U1 = np.array([U0(x) for x in X ])
for i in range(N):
    U = np.linalg.solve(M,U+tau*g3(U))
    U1 = np.linalg.solve(M,U1+tau*g2(U1))


plt.plot(X,U,"--", label='$u(1-u^2)$')    
plt.plot(X,U1,"b--+", label='u(1-u)')
plt.xlabel('x')
plt.title('T=1')
plt.ylabel('U')
plt.legend()
```
