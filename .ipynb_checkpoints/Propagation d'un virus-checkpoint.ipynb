# Propagation d'un virus 

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
import random
```

  **Mod√®le proportionnel**

```python
def trajectoire(X_0,m,alpha,p=1/5):
    def U_n(n,p):
        P=[1-p,p]
        U=np.random.choice([0,1],n,p=P)
        return U
    X=[X_0]
    for i in range (0,m-1):
        U=U_n(m,p)[i]
        X.append((1+alpha)**U*(1-alpha)**(i-U)*X_0)
    s=0
    for i in X:
        if i <X_0:
            s+=1
        s=s/len(X)
    print('P( X < X0)=', s)
    return X
```

```python
alpha = 0.5
n=40
p=0.5
X=trajectoire(30,n,alpha,p)
X1=trajectoire(50,n,alpha,p)
X2=trajectoire(80,n,alpha,p)
plt.xlabel('L\'instant t')
plt.ylabel('Nmbre de personnes malades √† t')
plt.plot(np.arange(n),X,label='X_0=30')
plt.plot(np.arange(n),X1,label='X_0=50')
plt.plot(np.arange(n),X2,label='X_0=80')
plt.legend()
```
