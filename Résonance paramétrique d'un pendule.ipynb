```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#RK4
def theta(w_,eps_):
    f=lambda t,U : np.array([U[1],-w_**2*(1+eps_*np.cos(t))*U[0]])
    V=np.array([np.pi/8,0])
    T=200
    dt=0.1
    N=int(T/dt)
    l=[V[0]]
    for i in range (N):
        k1=f(i*dt,V)
        k2=f(i*dt+dt/3,V+dt*k1/3)
        k3=f(i*dt+2*dt/3,V-dt*k1/3+dt*k2)
        k4=f((i+1)*dt,V+dt*k1-dt*k2+dt*k3)
        V=V+dt/8*(k1+3*k2+3*k3+k4)
        l.append(V[0])
    return l
```

```python
w1,eps1=1/2,0.1
w2,eps2=1,0.2
T=200
dt=0.1
N=int(T/dt)
plt.plot([i*dt for i in range (N+1)],theta(w1,eps1))
plt.title (' =1/2 et =0.1')
plt.xlabel('t')
plt.ylabel('$\Theta$')
```

```python
plt.plot([i*dt for i in range (N+1)],theta(w2,eps2))
plt.title (' =1 et =0.2')
plt.xlabel('t')
plt.ylabel('$\Theta$')
```
